```python
import numpy as np
import matplotlib.pyplot as plt
from module._rc_operations import reservoir, cycle
from module.construct_q import Q_mtx_fast
from module.fourier_motif import fourier_basis_motif, find_fourier_motifs
from module.util import sort_columns_by_partial_argmax, normalize_columns
```

```python
spectral_radius = 1
# Step 1: Define parameters for the refined Fourier basis
n_res = 97  # Length of each Fourier basis vector
interval_len = n_res * 3 # Ensure interval_len matches n_res

motif_w_rel_threshold = 0.0

normalize_flag = True
```

```python
W = cycle(n_res, spectral_radius = 1)

_, V = reservoir(n_res,1, rin = 1, spectral_radius = spectral_radius)

```

```python
Qp = Q_mtx_fast(W, V, interval_len)

#print('Computing motifs...')
motif, motif_weight, _ = np.linalg.svd(Qp, full_matrices=True)

# Filter motifs 
ind = np.where(motif_weight >= max(motif_weight) * motif_w_rel_threshold)[0]
filtered_motifs = motif[:, ind]

# Step 2: Track original indices before sorting
original_indices = np.arange(filtered_motifs.shape[1])

# Step 3: Sort the filtered motifs and keep track of sorted indices
filtered_motifs_fft = np.fft.fft(filtered_motifs, axis=0)
sorted_indices = np.argsort([np.argmax(np.abs(m[:m.size // 2])) for m in filtered_motifs_fft.T])
sorted_filtered_motifs = filtered_motifs[:, sorted_indices]
sorted_original_indices = original_indices[sorted_indices]

if normalize_flag:
    basis_norms = np.linalg.norm(sorted_filtered_motifs, axis=0)
    basis_norms[basis_norms == 0] = 1  # Avoid division by zero by setting zero norms to 1
    sorted_filtered_motifs = sorted_filtered_motifs / basis_norms

    
unsorted_fourier_basis, fourier_basis = find_fourier_motifs(W,V, motif_w_rel_threshold = 0, 
                                                            normalize_flag = False, interval_len = interval_len)

# Compute the FFTs
original_motif_fft = np.fft.fft(filtered_motifs, axis=0)
unsorted_fourier_basis_fft = np.fft.fft(unsorted_fourier_basis, axis=0)

# Step 6: Plot the FFTs
# plt.figure(figsize=(12, 6))

# Plot the FFT of the filtered motifs
# plt.subplot(1, 2, 1)

plt.figure(figsize=(6,6))
plt.imshow(normalize_columns(np.abs(original_motif_fft))[:,:n_res], aspect='auto', cmap='viridis')
plt.colorbar(label='Magnitude')
plt.title('Fourier Spectra of Motifs')
plt.xlabel('Motif Index')
plt.ylabel('$|FFT(m_k)(\omega)|^2$')
plt.tight_layout()

plt.savefig('results/spectra_motif.png')
plt.show()

# Plot the FFT of the corresponding Fourier basis
# plt.subplot(1, 2, 2)

plt.figure(figsize=(6,6))
plt.imshow(normalize_columns(np.abs(unsorted_fourier_basis_fft))[:,:n_res], aspect='auto', cmap='viridis')
plt.colorbar(label='Magnitude')
plt.title('Fourier Spectra of Closest Fourier Basis')
plt.xlabel('Fourier Basis Index')
plt.ylabel('$|FFT(m_k)(\omega)|^2$')

plt.tight_layout()
plt.savefig('results/spectra_fourier.png')
plt.show()

plt.figure(figsize=(12, 6))

# Plot the FFT of the filtered motifs
plt.subplot(1, 2, 1)
plt.imshow(filtered_motifs[:,:n_res], aspect='auto', cmap='viridis')
plt.colorbar(label='Magnitude')
plt.title('Motifs')
plt.xlabel('Motif Index')
plt.ylabel('$|FFT(m_k)(\omega)|^2$')

# Plot the FFT of the corresponding Fourier basis
plt.subplot(1, 2, 2)
plt.imshow(unsorted_fourier_basis[:,:n_res], aspect='auto', cmap='viridis')
plt.colorbar(label='Magnitude')
plt.title('Fourier Basis')
plt.xlabel('Fourier Basis Index')
plt.ylabel('$|FFT(m_k)(\omega)|^2$')

plt.tight_layout()
plt.show()
```
