```python
def custom_fft_with_basis(signal, N, basis_real=None, basis_imag=None):
    if basis_real is None or basis_imag is None:
        return np.fft.fft(signal)
    
    real_part = np.zeros(N)
    imag_part = np.zeros(N)
    
    for k in range(N):
        real_part[k] = np.dot(signal, basis_real[k])
        imag_part[k] = -np.dot(signal, basis_imag[k])
    
    return real_part + 1j * imag_part
```

```python
def normalize_signal(X):
    xm = X.mean()
    xs = X.std()
    if np.abs(xm - 0) < 1e-14 and np.abs(xs - 1) < 1e-14:
        normalize_flag = 0
        # print('Signal already normalized...')
    else:
        X = (X - X.mean()) / X.std()
        
    return X
```

```python
def custom_fft_with_basis2(signal, axis = 0, basis_real=None, basis_imag=None):
    if basis_real is None or basis_imag is None:
        return np.fft.fft(signal, axis=axis)

    N = basis_real.shape[0]  # Infer N from the basis

    # Initialize the result array
    shape = signal.shape
    if signal.ndim == 1:
        result = np.zeros((signal.shape[0], 1), dtype=complex)
    else:
        result = np.zeros(shape, dtype=complex)

    if axis == 0:
        for k in range(N):
            real_part = np.dot(signal, basis_real[k])
            imag_part = -np.dot(signal, basis_imag[k])
            result[k, :] = real_part + 1j * imag_part
    elif axis == 1:
        for k in range(N):
            real_part = np.dot(signal.T, basis_real[k])
            imag_part = -np.dot(signal.T, basis_imag[k])
            result[:, k] = real_part + 1j * imag_part
    else:
        raise ValueError("Axis must be 0 or 1")

    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal
N = 1024  # Number of samples
t = np.arange(N)
signal = np.sin(2 * np.pi * t / N * 5) + 0.5 * np.sin(2 * np.pi * t / N * 20)


# Define the DFT using custom basis vectors or default to numpy FFT
def custom_fft_with_basis(signal, axis = 0, basis_real=None, basis_imag=None):
    if basis_real is None or basis_imag is None:
        return np.fft.fft(signal, axis = axis)
    
    N = basis_real.shape[0]
    
    real_part = np.zeros(N)
    imag_part = np.zeros(N)
    
    if axis == 0:
        for k in range(N):
            real_part[k] = np.dot(signal, basis_real[k])
            imag_part[k] = -np.dot(signal, basis_imag[k])
    elif axis == 1:
        for k in range(N):
            real_part[k] = np.dot(signal.T, basis_real[k])
            imag_part[k] = -np.dot(signal.T, basis_imag[k])
    
    result = real_part + 1j * imag_part
    return result.ravel()

# Create the standard cosine and sine basis vectors
basis_real = 0.1*np.array([[np.cos(2 * np.pi * k * n / N) for n in range(N)] for k in range(N)])
basis_imag = 0.1*np.array([[np.sin(2 * np.pi * k * n / N) for n in range(N)] for k in range(N)])

# Compute the custom FFT with custom basis vectors
custom_fft_basis_result = custom_fft_with_basis(signal, axis = 0, basis_real=basis_real, basis_imag=basis_imag)

# Compute the FFT using the default numpy FFT (by not providing the basis)
default_fft_result = custom_fft_with_basis(signal, axis = 0)

# Verify the results
print("Are the custom FFT with custom basis and numpy FFT results close?", np.allclose(custom_fft_basis_result, default_fft_result))

# Plot the results for comparison
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Custom FFT with Basis Vectors")
plt.plot(np.abs(custom_fft_basis_result))
plt.subplot(1, 2, 2)
plt.title("Default Numpy FFT")
plt.plot(np.abs(default_fft_result))
plt.show()
```
