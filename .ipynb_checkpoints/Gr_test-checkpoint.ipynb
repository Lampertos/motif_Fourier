```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
A = np.array([[1,0,1],[0,1,1]]).transpose()
```

```python
A
```

```python
D = np.eye(3)[:,:2]

# D = np.eye(n)[:,:k]
```

```python
D
```

```python
A @ D.T
```

```python
B = A @ A.T
```

```python
B
```

```python


# Generating sample data
G = nx.from_numpy_array(B)
G.edges(data=True)
adjacency_matrix = nx.adjacency_matrix(G)

# The actual work
# You may prefer `nx.from_numpy_matrix`.
# G2 = nx.from_scipy_sparse_matrix(adjacency_matrix)
nx.draw_circular(G)
plt.axis('equal')
```

```python
#
# Initialize the latent directed connectivity matrix C with zeros
B = A
n = B.shape[0]
C = np.zeros((n, n))

# Compute the row-wise dot products to fill the directed connectivity matrix C
for i in range(n):
    for j in range(i + 1, n):  # Only for j > i to get upper triangular
        C[i, j] = np.dot(B[i], B[j].T)


# Display the resulting matrix C
C
```

```python


# Generating sample data
G = nx.from_numpy_array(C)
G.edges(data=True)
adjacency_matrix = nx.adjacency_matrix(G)

# The actual work
# You may prefer `nx.from_numpy_matrix`.
# G2 = nx.from_scipy_sparse_matrix(adjacency_matrix)
nx.draw_circular(G)
plt.axis('equal')
```

```python
W = np.diag([1, 2])  # Different weights for each component

B = A
# Initialize the latent directed connectivity matrix C with zeros
n = B.shape[0]
C = np.zeros((n, n))

# Compute the weighted row-wise dot products to fill the directed connectivity matrix C
for i in range(n):
    for j in range(n):
        C[i, j] = np.dot(B[i], (W @ B[j]).T)  # Weighted dot product for asymmetry

# Display the resulting non-symmetric matrix C
C
```

```python


# Generating sample data
G = nx.from_numpy_array(C)
G.edges(data=True)
adjacency_matrix = nx.adjacency_matrix(G)

# The actual work
# You may prefer `nx.from_numpy_matrix`.
# G2 = nx.from_scipy_sparse_matrix(adjacency_matrix)
nx.draw_circular(G)
plt.axis('equal')
```
