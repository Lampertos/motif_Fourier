```python
import numpy as np
import matplotlib.pyplot as plt

def generate_column_93(n_res):
    '''
    Generate the 93rd column of the Fourier basis matrix analytically.
    '''
    k = 46  # Corresponding to the 93rd column (2 * 46 + 1)
    column_93 = np.array([np.sin(2 * np.pi * k * m / n_res) for m in range(n_res)])
    return column_93
```

```python

# Original and denser resolutions for comparison
n_res_original = 97
n_res_dense = 500  # Example: increase to 500 for denser sampling

# Generate columns at different resolutions
column_93_original = generate_column_93(n_res_original)
column_93_dense = generate_column_93(n_res_dense)

# Plot the results on the same scale
plt.figure(figsize=(12, 6))
plt.plot(np.linspace(0, 1, n_res_original), column_93_original, label='Original Sampling (n_res=97)', marker='o', alpha = 0.5)
plt.plot(np.linspace(0, 1, n_res_dense), column_93_dense, label='Denser Sampling (n_res=500)')

plt.title("Comparison of 93rd Column at Different Sampling Densities without scaling factor of $\sqrt(2 / n_res)$")
plt.xlabel("Normalized Sample Index")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
