```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sine_wave(freq, sampling_points):
    return np.sin(2 * np.pi * freq * np.linspace(0, 1, sampling_points))

high_frequency = 46
high_sampling_points = 500
low_sampling_points = 97  # Low sampling to show misalignment

high_sampling_wave = generate_sine_wave(high_frequency, high_sampling_points)
low_sampling_wave = generate_sine_wave(high_frequency, low_sampling_points)

```

```python
# Create three plots: high-frequency sine curve, low-frequency sampled sine curve, and overlay of low-sampling points on the original

plt.figure(figsize=(12, 10))

# Top plot: high-frequency sine curve
plt.subplot(3, 1, 1)
plt.plot(np.linspace(0, 1, high_sampling_points), high_sampling_wave, label='High Sampling (500 points)')
plt.title("High Frequency Sine Curve with Dense Sampling")
plt.xlabel("Normalized Sample Index")
plt.ylabel("Amplitude")
plt.legend(loc='upper right')
plt.grid(True)

# plt.savefig('results/sampling_harmonic_high.png')
# plt.tight_layout()
# plt.show()


# Middle plot: low-frequency sine curve
plt.subplot(3, 1, 2)
plt.plot(np.linspace(0, 1, low_sampling_points), low_sampling_wave, label='Low Sampling (97 = $n = \tau$ points)', marker='o', color ='red')
plt.title("Low Frequency Sampling of the Same Sine Curve")
plt.xlabel("Normalized Sample Index")
plt.ylabel("Amplitude")
plt.legend(loc='upper right')
plt.grid(True)
# plt.savefig('results/sampling_harmonic_low.png')
# plt.tight_layout()
# plt.show()


# Bottom plot: overlay of low-sampling points on the high-sampling curve
plt.subplot(3, 1, 3)
plt.plot(np.linspace(0, 1, high_sampling_points), high_sampling_wave, label='High Sampling (500 points)')
plt.scatter(np.linspace(0, 1, low_sampling_points), low_sampling_wave, color='red', label='Low Sampling Points', zorder=5)
plt.title("Low Sampling Points Overlaid on High Frequency Sine Curve")
plt.xlabel("Normalized Sample Index")
plt.ylabel("Amplitude")
plt.legend(loc='upper right')
plt.grid(True)

plt.savefig('results/sampling_harmonic_full.png')

# plt.savefig('results/sampling_harmonic_over.png')
plt.tight_layout()
plt.show()
```
